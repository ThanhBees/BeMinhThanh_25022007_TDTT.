```python
#bai4
h = int(input('nhap do cao tam giac:'))
while h < 0:
    h = int(input('nhap do cao tam giac:'))
for i in range(1, h + 1):
    spaces =  " " * (h - i)
    stars = "*" * (2 * i - 1)
    print(spaces + stars)
```
