```python
#w5a1
a, b = map(int, input('nhap 2 so a va b lan luot').split())
def Max_Of_Two(a, b):
    if a < b:
        return b
    else:
        return a
print(Max_Of_Two(a, b))
```

```python
#w5a2
a, b = map(int, input('nhap 2 so a va b lan luot').split())
def swap(a, b):
    a, b = b, a
    return a, b
print(swap(a, b))
```

```python
#w5a3
a = int(input('nhap vao 1 so nguyen'))
def Test(x):
    flag = True
    if x < 2:
        flag = False
    else:
        for i in range(2, int(x**0.5) + 1):
            if x % i == 0:
                flag = False
                break
    return flag
if Test(a):
    print('la so nguyen to')
else:
    print('khong la so nguyen to')
```

```python
#w5a4
a = int(input('nhap vao 1 so nguyen'))
def Perfect_Num(x):
    s = 0
    for i in range(1 , x):
        if x % i == 0:
            s += i
    if s == x:
        return True
    else:
        return False
if Perfect_Num(a):
    print('la so hoan hao')
else:
    print('la so khong hoan hao')
```

```python
#w5a6
n = int(input('nhap vao so nguyen n:'))
def Giai_Thua(x):
    s = 1
    for i in range(1, x + 1):
        s *= i
    return s
print(Giai_Thua(n))
        
```

```python
#w5a7
a, b = map(float, input('nhap 2 so a va b').split())
operat = input('nhap phep toan') #nhap 1 trong 4 toan tu (+ - * /)
def May_Tinh(x, y):
    if operat == "+":
        return round(x + y, 2)
    elif operat == "-":
        return round(x - y, 2)
    elif operat == "*":
        return round(x * y, 2)
    elif operat == "/":
        return round(x / y, 2)
    else:
        return False
if May_Tinh(a, b):
    print(May_Tinh(a, b))
else:
    print('toan tu khong hop le')
```
